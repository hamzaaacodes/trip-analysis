```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_excel('./xlsx (1).xlsx')
dataset.head(3)
```

```python
plt.plot(dataset.Longitude, dataset.Latitude)
```

```python
dataset.Status.unique()
```

```python
ignition_on = dataset[dataset.Status == 'Ignition On']
ignition_on.head()
```

```python
ignition_off = dataset[dataset.Status == 'Ignition Off']
ignition_off.head()
```

From above two DF, we can conclude that the first trip starts with loc 18 and ends with 408.

```python
trip_1 = dataset.loc[18:408: ]
trip_1
```

```python
plt.plot(trip_1['Longitude'], trip_1['Latitude'], linestyle='dashed', color='green')
```

* The above line graph represents the whole trip w.r.t longitude and latitude.

```python
over_speeding = trip_1[trip_1['Speed(Km/h)'] > 80]
over_speeding
```

```python
plt.plot(trip_1['Longitude'], trip_1['Latitude'], linestyle='dashed', color='green')
plt.plot(over_speeding['Longitude'], over_speeding['Latitude'], color='red', linewidth=3)
```

* The above line graph represents the whole trip w.r.t longitude and latitude.
* The red line represents, where overspeeding occured.

```python
seatbelt_on = trip_1[trip_1.Status == 'Driver SeatBelt Open']
seatbelt_on
```

```python
seatbelt_off = trip_1[trip_1.Status == 'Driver SeatBelt Close']
seatbelt_off
```

```python
plt.plot(trip_1['Longitude'], trip_1['Latitude'], linestyle='dashed', color='green')
plt.plot(over_speeding['Longitude'], over_speeding['Latitude'], color='red', linewidth=3)

plt.scatter(seatbelt_on['Longitude'], seatbelt_on['Latitude'], color='brown')
plt.scatter(seatbelt_off['Longitude'], seatbelt_off['Latitude'], color='black')
```

* The above line graph represents the whole trip w.r.t longitude and latitude.
* The red line represents, where overspeeding occured.
* The brown dot represents, when the driver put on the seat belt.
* The black dot represents, when the driver off on the seat belt.

```python
harsh_breaks = abs(trip_1['Speed(Km/h)'].diff())
trip_1['Harsh_breaks'] = harsh_breaks
trip_1
```

```python
harsh_break_point = trip_1[trip_1['Harsh_breaks'] > 30]
harsh_break_point
```

```python
plt.plot(trip_1['Longitude'], trip_1['Latitude'], linestyle='dashed', color='green')
plt.plot(over_speeding['Longitude'], over_speeding['Latitude'], color='red', linewidth=3)

plt.scatter(seatbelt_on['Longitude'], seatbelt_on['Latitude'], color='brown')
plt.scatter(seatbelt_off['Longitude'], seatbelt_off['Latitude'], color='black')

plt.scatter(harsh_break_point['Longitude'], harsh_break_point['Latitude'], color='red')
```

* The above line graph represents the whole trip w.r.t longitude and latitude.
* The red line represents, where overspeeding occured.
* The brown dot represents, when the driver put on the seat belt.
* The black dot represents, when the driver off on the seat belt.
* The red dot represents, when the driver breaks harshly

```python
max_speed = trip_1['Speed(Km/h)'].max()
max_speed_points = trip_1[trip_1['Speed(Km/h)'] == max_speed]
max_speed_points
```

```python
plt.plot(trip_1['Longitude'], trip_1['Latitude'], linestyle='dashed', color='green')
plt.plot(over_speeding['Longitude'], over_speeding['Latitude'], color='red', linewidth=3)

plt.scatter(seatbelt_on['Longitude'], seatbelt_on['Latitude'], color='brown')
plt.scatter(seatbelt_off['Longitude'], seatbelt_off['Latitude'], color='black')

plt.scatter(harsh_break_point['Longitude'], harsh_break_point['Latitude'], color='red')

plt.scatter(max_speed_points['Longitude'], max_speed_points['Latitude'], color='blue')

plt.show()
```

* The above line graph represents the whole trip w.r.t longitude and latitude.
* The red line represents, where overspeeding occured.
* The brown dot represents, when the driver put on the seat belt.
* The black dot represents, when the driver off on the seat belt.
* The red dot represents, when the driver applied harsh breaks.
* The blue dots represents, when the driver was driving at maximum speed.
